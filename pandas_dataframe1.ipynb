```python
# DataFrame 생성하기
member = {'번호' : [1,2,3,4,5],
          '이름' : ['홍길동', '전우치', '강감찬', '일지매', '을지문덕'],
          '매출' : [100,200,300,250,150]}
member_1 = pd.DataFrame(member)
member_1
```

```python
# DataFrame 합치기(merge)
df_1 = {'학번': ['1001','1002','1003'],
        '이름': ['홍길동', '일지매', '전우치'],
        '키': [180, 175, 168]}
stu_1 = pd.DataFrame(df_1)
stu_1
```

```python
# DataFrame 합치기(merge) 데이터 생성2
df_2 = {'학번': ['1001','1002','1003'],
        '이름': ['홍길동', '일지매', '강감찬'],
        '몸무게': [80, 75, 68]}
stu_2 = pd.DataFrame(df_2)
stu_2
```

```python
# 기본값으로 merge 수행 - inner join 방식
pd.merge(stu_1, stu_2, left_on='이름', right_on='이름')
```

```python
# merge 예제 - merge 할 기준을 지정하기(left join)
pd.merge(stu_1, stu_2, how='left', left_on='이름', right_on='이름')
```

```python
# merge 예제 - merge 할 기준을 지정하기(right join)
pd.merge(stu_1, stu_2, how='right', left_on='이름', right_on='이름')
```

```python
# merge 예제 - merge 할 기준을 지정하기(outer join)
pd.merge(stu_1, stu_2, how='outer', left_on='이름', right_on='이름')
```

```python
# concat (axis=0, 세로로 합치기)
pd.concat([stu_1, stu_2], axis=0)
```

```python
# concat (axis=0, 가로로 합치기)
pd.concat([stu_1, stu_2], axis=1)
```
